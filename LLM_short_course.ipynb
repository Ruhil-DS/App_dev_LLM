```python
def get_completion(prompt, model='gpt-3.5-turbo'):
    msg = [{"role": 'user', "content": prompt}]
    resp = openai.chat.completions.create(
        model=model,
        messages=msg,
        temperature=0
    )
    return resp.choices[0].message['content']
```
